```python

import os
import zipfile
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt

from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split

zip_path = "/content/flood-area-segmentation.zip"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("/content")

print("Dataset Extracted")


print(os.listdir('/content'))
print("Images:", len(os.listdir('/content/Image')))
print("Masks:", len(os.listdir('/content/Mask')))


image_dir = "/content/Image"
mask_dir = "/content/Mask"

IMG_SIZE = 128


images = []
masks = []

image_files = sorted(os.listdir(image_dir))
mask_files = sorted(os.listdir(mask_dir))

for img_file, mask_file in zip(image_files, mask_files):

    img = cv2.imread(os.path.join(image_dir, img_file))
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img / 255.0

    mask = cv2.imread(os.path.join(mask_dir, mask_file), 0)
    mask = cv2.resize(mask, (IMG_SIZE, IMG_SIZE))
    mask = mask / 255.0
    mask = np.expand_dims(mask, axis=-1)

    images.append(img)
    masks.append(mask)

images = np.array(images)
masks = np.array(masks)

print("Images Shape:", images.shape)
print("Masks Shape:", masks.shape)

X_train, X_test, y_train, y_test = train_test_split(
    images, masks, test_size=0.2, random_state=42
)


def unet_model():

    inputs = layers.Input((IMG_SIZE, IMG_SIZE, 3))

    c1 = layers.Conv2D(16,3,activation='relu',padding='same')(inputs)
    c1 = layers.Conv2D(16,3,activation='relu',padding='same')(c1)
    p1 = layers.MaxPooling2D()(c1)

    c2 = layers.Conv2D(32,3,activation='relu',padding='same')(p1)
    c2 = layers.Conv2D(32,3,activation='relu',padding='same')(c2)
    p2 = layers.MaxPooling2D()(c2)

    c3 = layers.Conv2D(64,3,activation='relu',padding='same')(p2)
    c3 = layers.Conv2D(64,3,activation='relu',padding='same')(c3)

    u4 = layers.UpSampling2D()(c3)
    u4 = layers.concatenate([u4,c2])

    c4 = layers.Conv2D(32,3,activation='relu',padding='same')(u4)
    c4 = layers.Conv2D(32,3,activation='relu',padding='same')(c4)

    u5 = layers.UpSampling2D()(c4)
    u5 = layers.concatenate([u5,c1])

    c5 = layers.Conv2D(16,3,activation='relu',padding='same')(u5)
    c5 = layers.Conv2D(16,3,activation='relu',padding='same')(c5)

    outputs = layers.Conv2D(1,1,activation='sigmoid')(c5)

    model = models.Model(inputs, outputs)

    return model

model = unet_model()

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()


history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=10,
    batch_size=8
)

prediction = model.predict(X_test[:1])

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(X_test[0])
plt.title("Satellite Image")

plt.subplot(1,3,2)
plt.imshow(y_test[0].squeeze(), cmap='gray')
plt.title("Actual Flood Mask")

plt.subplot(1,3,3)
plt.imshow(prediction[0].squeeze(), cmap='gray')
plt.title("Predicted Flood Area")

plt.show()
```
